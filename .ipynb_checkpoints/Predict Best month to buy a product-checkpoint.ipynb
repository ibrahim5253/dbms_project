```python
# Dataset used (http://jmcauley.ucsd.edu/data/amazon/)
# Dataset is 5-core i.e. each product and user has at least 5 reviews 
# 1.reviews_Electronics 
# 2.reviews_Health_and_Personal_Care
# 3.reviews_Clothing_Shoes_and_Jewelry
# 4.reviews_Cell_Phones_and_Accessories

########################################################################
# Format of the Data
# reviewerID - ID of the reviewer, e.g. A2SUAM1J3GNN3B
# asin - ID of the product, e.g. 0000013714
# reviewerName - name of the reviewer
# helpful - helpfulness rating of the review, e.g. 2/3
# reviewText - text of the review
# overall - rating of the product
# summary - summary of the review
# unixReviewTime - time of the review (unix time)
# reviewTime - time of the review (raw)
########################################################################


# Assumption : 
# 1. The time when review was written is within a 
#    month's neigbhourhood of purchase of the product
```

```python
import pandas as pd
import numpy as np
import gzip
import json  
from urllib.parse import urlencode
from urllib.request import Request, urlopen
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords
```

```python
def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield eval(l)
```

```python
def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')
```

```python
datasets = {'cellphone':'reviews_Cell_Phones_and_Accessories_5.json.gz', 
            'clothing': 'reviews_Clothing_Shoes_and_Jewelry_5.json.gz',
            'health':'reviews_Health_and_Personal_Care_5.json.gz',
            'electronics':'reviews_Electronics_5.json.gz'
           }
```

```python
# load dataset into DataFrame
df = getDF(datasets['clothing'])
df
```

```python
# define: get GALC Emotional score for all 39 class
def getGALCEmotionalScore(reviewtext, class_map, classnames):
    
    result = {cname:0.0 for cname in classnames}
    stop = set(stopwords.words('english'))
    wordlist = [i for i in reviewtext.lower().split() if i not in stop]
               
    stemmer = SnowballStemmer("english")
    emotional_words=0.0
    for word in wordlist:
        if word in class_map:
            print(word)
            result[class_map[word]] += 1.0
            emotional_words+=1.0
        elif stemmer.stem(word) in class_map:
            print(word)
            result[class_map[stemmer.stem(word)]] += 1.0
            emotional_words+=1.0
   
    # Normalise in case a person is too enthusiastic 
    #for cname in result.keys():
        #result[cname] /= float(emotional_words)
    return result
```

```python
def getSentimentScoreTextProcessingAPI(reviewText, summary):
    url = 'http://text-processing.com/api/sentiment/' 
    post_fields = {'text': reviewText}     
    request = Request(url, urlencode(post_fields).encode())
    jsonobj = urlopen(request).read().decode()
    print(jsonobj)
    jsonobj = json.loads(jsonobj)
    if jsonobj["label"] == "neutral":
        return 0
    else:    
        results = jsonobj["probability"]
        neg, pos, neu =  results["neg"], results["pos"], results["neutral"]
        if pos > neg:
            return pos-neg
        else :
            return -(neg-pos)


# Set destination URL here
# Set POST fields here
def getSentimentScore(reviewText, summary):
    url = 'http://sentiment.vivekn.com/api/text/' 
    post_fields = {'txt': reviewText}     
    request = Request(url, urlencode(post_fields).encode())
    jsonobj = urlopen(request).read().decode()
    print(jsonobj)
    jsonobj = json.loads(jsonobj)
    result = jsonobj["result"]
    if result['sentiment'] == "Positive":
        return float(result['confidence'])/100.0
    elif result['sentiment'] == "Negative":
        return (-1)*float(result['confidence'])/100.0
    else:
        return 0

    
def getSentimentScoreStanFordCoreNLP(reviewText, summary):
    url = 'http://sentiment.vivekn.com/api/text/' 
    post_fields = {'txt': reviewText}     
    request = Request(url, urlencode(post_fields).encode())
    jsonobj = urlopen(request).read().decode()
    print(jsonobj)
    jsonobj = json.loads(jsonobj)
    result = jsonobj["result"]
    if result['sentiment'] == "Positive":
        return float(result['confidence'])/100.0
    elif result['sentiment'] == "Negative":
        return (-1)*float(result['confidence'])/100.0
    else:
        return 0
```

```python
# 1. Map (Emotions) score of each proucts' review to it's Time(a  month)
# 2  Give score for each month (impact of emotional content on influence) based
#    2.1 Helpfulness of the review (w = fraction poitive i.e (r+ve)/(r-ve + r+ve))
#        to give weight to sentiment, rating, emotinal scores 
#    2.2 Positive or Negative score of Review using Sentiment Analysis
#    2.3 Overall rating
#    2.4 Emotional Words(GALC defined for 39 classes)
#    2.5 Normalise over each month

# Model 
# Input :
# 1. dataset : dataframe of the dataset
# 2. w_helpful(minimum helpfulness), w_sentiment, w_rating, w_emotinal; 
#     for managing impact:
#     w_helpful + w_sentiment + w_rating = 1   
#     0 ≤ w_helpful , w_sentiment , w_rating ≤ 1
# Output :
# returns a dictionary with score info for each month for each product::
# return (product id, [score_list])  
def getscores(dataset, w_helpful, w_sentiment, w_rating, count):
    
    # Load GALC lexicon
    galc = np.genfromtxt(fname='./English-Table 1.csv',delimiter=",",dtype=str)

    # PreProcessing on Lexicon
    # 1.Map all words in lexicon to their respective classes 
    #  using dictionary
    # 2. Store all class names in a list
    class_map, classnames = {}, set()
    for i in range(len(galc)):
        for j in range(len(galc[i])):
                if j != 0 and len(galc[i,j]) > 0:
                    if str(galc[i,j][-1:]) == "*":
                        class_map[galc[i,j][:-1]]=galc[i,0]
                    else:
                        class_map[galc[i,j]]=galc[i,0]
        classnames.add(galc[i,0])
    print(class_map.keys())
    
    # process each review and store aggregated results in 'results'
    results = {}
    #dict format :: (product id, [score_list] or [count_list]) 
    sentiment, rating, emotion_score, total = {}, {}, {}, {}
    
    # Extract month for each review
    it = 0
    for index, row in df.iterrows():
        it+=1
        # extract month and Product Id    
        month, asin = int(row['reviewTime'][0:3]), row['asin']
        
        #initialise values in dictionary for a product
        if asin not in total:
            total[asin], rating[asin] = [0]*12, [0]*12
            sentiment[asin], emotion_score[asin] =  [0]*12, [{cname:0 for cname in classnames} for i in range(12)] 
            
        # Find The Helpfulness score of a review:
        # if only -ve votes than review is not helpful with ww_helpful=0
        # else if only +ve votes than review is helpful with ww_helpful=1
        # else if both votes than calculate fraction (+ve/(+ve + -ve))
        helpfulness = row['helpful']
        ww_helpful = w_helpful
        
        if helpfulness[0] != 0 or helpfulness[1] !=0:
            if helpfulness[0] !=0 and helpfulness[1] !=0:
                ww_helpful = helpfulness[0]/(helpfulness[0]+helpfulness[1])
            if helpfulness[1] == 0:
                ww_helpful[0]=1    
        # increase the count of review for that month
        if ww_helpful > 0:
            total[asin][month-1] += 1
            
        # update the rating score for that month
        rating[asin][month-1] += ww_helpful * w_rating * float(row['overall'])
        
        # Check Sentiment Score (i.e opinion, polarity)
        sentiment[asin][month-1] += ww_helpful * w_sentiment * getSentimentScoreTextProcessingAPI(row['reviewText'], row['summary'])
        
        # Emotional class scores (G.A.l.C)
        tmp = getGALCEmotionalScore(row['reviewText'], class_map, classnames)
        #print(tmp)
        for cname in classnames:
            emotion_score[asin][month-1][cname] += tmp[cname]
        if it >= count:
            break
        
    # Normalise scores
    it=0
    for key, values in total.items():
        it+=1
        results[key] = [0]*12
        for m in range(12):
            if total[key][m]!=0:
                results[key][m] += rating[key][m]/total[key][m]
                results[key][m] += sentiment[key][m]/total[key][m]
                results[key][m] /= 2.0
        if it >= count:
            break
    return results, emotion_score
```

```python
result, emotion_score = getscores(df,0.25,0.25,0.25,800)
```

```python
tmp_w, tmp_b = {}, {}
month_w, month_b = [0 for i in range(12)], [0 for i in range(12)]
for productid in result:
    min_v,min_m,max_v,max_m = result[productid][0],  0, result[productid][0], 0
    for month in range(1,12):
        if (min_v > result[productid][month] and result[productid][month] > 0) or min_v == 0:
            min_v = result[productid][month]
            min_m = month
        if max_v < result[productid][month]:
            max_v = result[productid][month]
            max_m = month
    month_w[min_m] +=1 
    month_b[max_m] +=2
    for em_class in emotion_score[productid][max_m]:
        if em_class not in tmp_b:
            tmp_b[em_class] = 0
        else:    
            tmp_b[em_class] += emotion_score[productid][max_m][em_class]
    for em_class in emotion_score[productid][min_m]:
        if em_class not in tmp_w:
            tmp_w[em_class] = 0
        else:    
            tmp_w[em_class] += emotion_score[productid][min_m][em_class]
```

```python
sum_w,sum_b = 0,0
for v in tmp_w.values():
    sum_w += v
for v in tmp_b.values():
    sum_b += v
tmp_w
```
